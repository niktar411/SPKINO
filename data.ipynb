```python
import requests

def get_movie_info_omdb(api_key: str, title: str = None, imdb_id: str = None, plot: str = "full") -> dict:

    base_url = "http://www.omdbapi.com/"
    params = {
        "apikey": api_key,  # Ключ API передаём здесь
        "plot": plot,
        "r": "json"
    }

    if title:
        params["t"] = title
    if imdb_id:
        params["i"] = imdb_id

    response = requests.get(base_url, params=params)

    if response.status_code != 200:
        print(f"Ошибка запроса: {response.status_code}")
        return {}

    data = response.json()
    if data.get("Response") == "False":
        print(f"Ошибка API: {data.get('Error')}")
        return {}

    return data
```

```python
# Пример использования
API_KEY = "20580e65"        # Твой API ключ без префиксов
IMDB_ID = "tt3896198"       # IMDb ID фильма

data = get_movie_info_omdb(api_key=API_KEY, imdb_id=IMDB_ID)
print(data)
```
